```python
import numpy as np
import matplotlib.pyplot as plt

# Thng s tn hiu
Fs = 100  # Tn s ly mu (Hz)
T = 1     # Thi gian quan st (giy)
t = np.linspace(0, T, Fs, endpoint=False)  # Trc thi gian
f = 5     # Tn s ca tn hiu (Hz)

# Tn hiu sin
x = np.sin(2 * np.pi * f * t)

# Bin i Fourier (DFT)
X_f = np.fft.fft(x)  # Tnh DFT ca tn hiu
frequencies = np.fft.fftfreq(Fs, 1/Fs)  # Trc tn s

# Ch ly mt na ph (do tnh i xng)
half_range = Fs // 2
X_magnitude = np.abs(X_f[:half_range])  # Ly bin  ph
frequencies_half = frequencies[:half_range]  # Ly trc tn s dng

# V tn hiu gc
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(t, x, label="x(t) = sin(25t)")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.title("Original Signal")
plt.legend()
plt.grid()

# V Ph bin 
plt.subplot(1, 2, 2)
plt.stem(frequencies_half, X_magnitude)
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.title("Fourier Transform (Magnitude Spectrum)")
plt.grid()

plt.tight_layout()
plt.show()
```
