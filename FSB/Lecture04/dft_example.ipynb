```python
import numpy as np
import matplotlib.pyplot as plt

# Tạo tín hiệu rời rạc (sóng sin)
N = 16  # Số mẫu
n = np.arange(N)
x = np.sin(2 * np.pi * n / N)  # Sóng sin

# Tính DFT bằng numpy
X = np.fft.fft(x)  

# Vẽ tín hiệu gốc và phổ tần số
plt.figure(figsize=(10, 4))

# Vẽ tín hiệu gốc (miền thời gian)
plt.subplot(1, 2, 1)
plt.stem(n, x)
plt.title("Tín hiệu gốc (Miền thời gian)")
plt.xlabel("n")
plt.ylabel("x[n]")

# Vẽ biên độ phổ (miền tần số)
plt.subplot(1, 2, 2)
plt.stem(np.abs(X))
plt.title("Biên độ phổ (Miền tần số)")
plt.xlabel("k")
plt.ylabel("|X[k]|")

plt.tight_layout()
plt.show()
```
