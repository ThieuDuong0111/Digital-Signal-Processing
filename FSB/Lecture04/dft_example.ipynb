```python
import numpy as np
import matplotlib.pyplot as plt

# To tn hiu ri rc (sng sin)
N = 16  # S mu
n = np.arange(N)
x = np.sin(2 * np.pi * n / N)  # Sng sin

# Tnh DFT bng numpy
X = np.fft.fft(x)  

# V tn hiu gc v ph tn s
plt.figure(figsize=(10, 4))

# V tn hiu gc (min thi gian)
plt.subplot(1, 2, 1)
plt.stem(n, x)
plt.title("Tn hiu gc (Min thi gian)")
plt.xlabel("n")
plt.ylabel("x[n]")

# V bin  ph (min tn s)
plt.subplot(1, 2, 2)
plt.stem(np.abs(X))
plt.title("Bin  ph (Min tn s)")
plt.xlabel("k")
plt.ylabel("|X[k]|")

plt.tight_layout()
plt.show()
```
